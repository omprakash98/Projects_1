```python
cols_to_use = ['loan_status','Principal','terms','past_due_days','age','education','Gender']
df1 = df0[cols_to_use]
df1
```

```python
df1['Principal'].unique()
```

```python
df2['paid_on_time'] = df0['loan_status'].apply(lambda x: 1 if x=='PAIDOFF' else 0)
df2 = df2.drop(['loan_status'],axis=1)
df2.head()
```

```python
# df2['gender']=df2['Gender'].apply(lambda x: 1 if x == 'female' else 0)
# df2.drop(['Gender'],axis = 1,inplace =True)
# df2.head()

df2['past_due_days'] = df2['past_due_days'].fillna(0)
df2
```

```python
df2_dummies= pd.get_dummies(df2[['education','Gender']],drop_first= True)
print(type(df2_dummies))
```

```python
df3 = pd.concat([df2,df2_dummies],axis = 1)
df3.drop(['education','Gender','past_due_days'],axis = 1,inplace =True)
df3.head()
#male =1 ,female =0,
```

```python
input_data =df3.drop(['paid_on_time'],axis = 1)
target = df3['paid_on_time']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input_data,target, test_size=0.2)
```

```python
len(X_train)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C=1)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict_proba([[1000,30,45,1,1,0,1]])
```

```python
model.predict_proba(input_data)
```
